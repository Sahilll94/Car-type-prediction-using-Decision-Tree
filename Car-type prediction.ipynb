```python
#Lets import scikit learn
import sklearn
```

```python
from sklearn import tree
#Tree is the module which provids decision tree classifier
```

```python
#features are number of seats and horse power
features = [[400, 2], [500, 2], [100.6, 4], [150, 6], [50, 4]]
labels = ["Super-car", "Super-car", "Hatchback", "SUV", "Hatchback"]


#manual prediction for [450,2]
```

```python
#initialising it or instantiating the decision tree classfier or DTC
classifier=tree.DecisionTreeClassifier()
```

```python
#Feed our data to the ML model
#Will use "FIT"
#model.fit(data)
```

```python
classifier = classifier.fit(features,labels)
#Fitting the model and feeding it in the data
```

```python
#to make predictions we use the .predict() method
#trained_model.predict(inputData)
```

```python
#prediction=classifier.predict([[HP,No of seat]])
prediction=classifier.predict([[400,2]]) #here we will put the HP and no of seat that we want to make prediction
print("This car type is:",prediction) #Concatenation
```

# Here, we will take various horse power and seat as input and decision tree will predict the car type

```python
#prediction=classifier.predict([[HP,No of seat]])
prediction=classifier.predict([[1000,1]]) #here we will put the HP and no of seat that we want to make prediction
print("This car type is:",prediction) #Concatenation
```

```python
#prediction=classifier.predict([[HP,No of seat]])
prediction=classifier.predict([[0,1]]) #here we will put the HP and no of seat that we want to make prediction
print("This car type is:",prediction) #Concatenation
```

```python
#prediction=classifier.predict([[HP,No of seat]])
prediction=classifier.predict([[1000,5]]) #here we will put the HP and no of seat that we want to make prediction
print("This car type is:",prediction) #Concatenation
```
